## Aggregation

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [5, 4, 3, 2, 1]
})

res=df.aggregate({'A':['sum','min'],'B' : ['max','mean'],'C' :['max','median']})
```

```python
res
```

## Group By 

```python
import pandas as pd

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'A','D'],
    'Sales': [100, 200, 150, 300, 250, 50, 400, 100, 900]
}
df=pd.DataFrame(data)

res=df.groupby('Category')['Sales'].sum()

print(res)
```

```python
df['Year'] = [2020, 2020, 2021, 2020, 2021, 2021, 2020, 2021,2020]
res=df.groupby(['Category','Year'])['Sales'].sum()
res
```
